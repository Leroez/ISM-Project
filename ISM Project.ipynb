```python
import pandas as pd
#change file path to wherever you have the files
path = r'//Users/benrivera/Documents/School/ISM 210/Project/Project/Employees.xlsx'
employeeData = pd.read_excel(path)
path2 = r'/Users/benrivera/Documents/School/ISM 210/Project/Project/Orders.xlsx'
orderData = pd.read_excel(path2)
```

Data Preparation

```python
#this merges the two dataframes using EmplyeeID
mergedData= (pd.merge(employeeData, orderData, left_on='EmployeeID', right_on='EmployeeID', how='outer'))
#resets the index
mergedData.reset_index()

mergedData
```

```python
#changes the data type, whether it is a string, int, or date and time
mergedData.LastName = mergedData.LastName.astype('string')

mergedData.FirstName = mergedData.FirstName.astype('string')

mergedData.Title = mergedData.Title.astype('string')

mergedData.TitleOfCourtesy = mergedData.TitleOfCourtesy.astype('string')

mergedData.Address = mergedData.Address.astype('string')

mergedData.City_x = mergedData.City_x.astype('string')

mergedData.PostalCode= mergedData.PostalCode.astype('string')

mergedData.Country_x = mergedData.Country_x.astype('string')

mergedData.HomePhone = mergedData.HomePhone.astype('string')

mergedData['Order ID']= mergedData['Order ID'].astype('string')

mergedData['Ship Mode'] = mergedData['Ship Mode'].astype('string')

mergedData['Customer ID'] = mergedData['Customer ID'].astype('string')

mergedData['Customer Name'] = mergedData['Customer Name'].astype('string')

mergedData['Segment'] = mergedData['Segment'].astype('string')

mergedData['Country_y'] = mergedData['Country_y'].astype('string')

mergedData['City_y'] = mergedData['City_y'].astype('string')

mergedData['State'] = mergedData['State'].astype('string')

mergedData['Region'] = mergedData['Region'].astype('string')

mergedData['Product ID'] = mergedData['Product ID'].astype('string')

mergedData['Category'] = mergedData['Category'].astype('string')

mergedData['Sub-Category'] = mergedData['Sub-Category'].astype('string')

mergedData['Product Name'] = mergedData['Product Name'].astype('string')

#Shows the data types
mergedData.dtypes
```

```python
#Imports a dependency
import numpy as np 
#This saves a column to a variable
QuantData = mergedData['Quantity']
#Saves the mean fo Quantity to a varaible
QuantMean = mergedData['Quantity'].mean()
#This finds the standard deviation of the quantity
QuantSD = mergedData['Quantity'].std()
#This number is the threshold
QuantThreshold = QuantMean + (3*QuantSD)
#shows which values exceed the threshold
QuantData[(np.abs(QuantData) > QuantThreshold)]
```

```python
import numpy as np 
#This saves a column to a variable
SalesData = mergedData['Sales']
#Saves the mean fo Sales to a varaible
SalesMean = mergedData['Sales'].mean()
#This finds the standard deviation of the sales
SalesSD = mergedData['Sales'].std()
#This number is the threshold
SalesThreshold = SalesMean + (3*SalesSD)
#shows which values exceed the threshold
SalesData[(np.abs(SalesData) > SalesThreshold)]
```

	Which sub-category is the most profitable in the State of California in 2015? 

```python
import numpy as np
def top(df, column, n=5):
    return df.sort_values(by=column)[-n:]
```

```python


filter= mergedData[mergedData['Order Date'].dt.year == 2015]
#Groups my data
Group1 = filter.groupby(['State','Sub-Category', filter['Order Date'].astype("datetime64[Y]")])

top_df = Group1.apply(top, column='Profit', n=1)
#sets index to sub-catagories
top_df.set_index('Sub-Category')
#Puts the sub-category and state up front and orders it based off of sales, also shows the most porfitable
top_df[(top_df['State']=="California")].sort_values(by='Profit', ascending=False).head(1)


```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#Groups my data
mergedDataYearly = mergedData.groupby(['State', mergedData['Order Date'].astype("datetime64[M]")])
#Sums up the data and resets the index
sumAGG = mergedDataYearly.sum().reset_index()
#Ensures we only see California
sumAGGNCAL = sumAGG[(sumAGG.State=="California")]
#resets index and drops older index
sumAGGNCAL.reset_index(drop=True)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#plots the sales
ax.plot(sumAGGNCAL.Sales, linestyle='--', color='g', marker='o', label='Sales')
#Sets the labels
labels = ax.set_xticklabels([0,'Jan 2015', 'Sep 2015', 'Jun 2016', 'April 2017', 'Jan 2018', 'Nov 2018'])
```
